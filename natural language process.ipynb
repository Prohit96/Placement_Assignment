```python
Q-2. Take any pdf and your task is to extract the text from that pdf and store it in a
csv file and then you need to find the most repeated word in that pdf.
```

```python
import csv
from collections import Counter
import PyPDF2

def extract_text_from_pdf(pdf_path):
    text = ""
    with open(pdf_path, "rb") as file:
        pdf = PyPDF2.PdfReader(file)
        num_pages = len(pdf.pages)
        for page_num in range(num_pages):
            page = pdf.pages[page_num]
            text += page.extract_text()
    return text

# Example usage
pdf_path = "DAX_Notes.pdf"
text = extract_text_from_pdf(pdf_path)
print(text)
csv_path = "output.csv"
with open(csv_path, 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Text'])
    writer.writerow([text])

    # Split the text into words
words = text.split()

# Count the occurrences of each word
word_counts = Counter(words)

# Find the most repeated word
most_common_word = word_counts.most_common(1)[0][0]

print("The most repeated word in the PDF is:", most_common_word)
```

```python
Q-1. Take any YouTube videos link and your task is to extract the comments from
that videos and store it in a csv file and then you need define what is most
demanding topic in that videos comment section.
```

```python
from googleapiclient.discovery import build
import csv
from collections import Counter
def extract_comments(video_id, api_key):
    api_service_name = "youtube"
    api_version = "v3"
    youtube = build(api_service_name, api_version, developerKey=api_key)

    comments = []

    # Get the video comments
    next_page_token = None
    while True:
        response = youtube.commentThreads().list(
            part="snippet",
            videoId=video_id,
            maxResults=100,
            pageToken=next_page_token
        ).execute()

        # Extract the comment text from the response
        for item in response["items"]:
            comment = item["snippet"]["topLevelComment"]["snippet"]["textDisplay"]
            comments.append(comment)

        # Check if there are more pages of comments
        if "nextPageToken" in response:
            next_page_token = response["nextPageToken"]
        else:
            break

    return comments
video_id = "gfDE2a7MKjA"
api_key = "AIzaSyAyBgKdNFO56tDfCXHzDmnwXx8_wU-XWUQ"

comments = extract_comments(video_id, api_key)

csv_path = "comments.csv"
with open(csv_path, 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Comment'])
    writer.writerows(zip(comments))

# Split the comments into words
words = ' '.join(comments).split()

# Count the occurrences of each word
word_counts = Counter(words)

# Find the most repeated word
most_common_word = word_counts.most_common(1)[0][0]

print("The most demanding topic in the video's comment section is:", most_common_word)
```

```python
Q-4. Take any text file and now your task is to Text Summarization without using
hugging transformer library
```

```python
import re
import heapq

def text_summarization(text, num_sentences):
    # Preprocess the text
    cleaned_text = re.sub(r'\n', ' ', text)  # Remove line breaks
    sentences = re.split(r'(?<=[.!?])\s+', cleaned_text)  # Tokenize into sentences

    # Calculate sentence scores
    word_frequencies = {}
    for sentence in sentences:
        words = re.findall(r'\b\w+\b', sentence.lower())  # Tokenize into words
        for word in words:
            if word not in word_frequencies:
                word_frequencies[word] = 1
            else:
                word_frequencies[word] += 1

    sentence_scores = {}
    for i, sentence in enumerate(sentences):
        words = re.findall(r'\b\w+\b', sentence.lower())  # Tokenize into words
        for word in words:
            if word in word_frequencies:
                if i not in sentence_scores:
                    sentence_scores[i] = word_frequencies[word]
                else:
                    sentence_scores[i] += word_frequencies[word]

    # Select the top-ranked sentences
    selected_sentences = heapq.nlargest(num_sentences, sentence_scores, key=sentence_scores.get)
    summary = ' '.join(sentences[i] for i in selected_sentences)

    return summary

# Read the text file
filename = r'input.txt'
with open(filename, 'r') as file:
    text = file.read()

# Set the desired number of sentences for the summary
num_sentences = 3

# Generate the summary
summary = text_summarization(text, num_sentences)

# Print the summary
print(summary)
```

```python
Q-3. from question 2, As you got the CSV and now you need perform key word
extraction from that csv file and do the Topic modeling
```

```python
import pandas as pd

# Load the CSV file
data = pd.read_csv('output.csv')

# Assuming the text column is named 'text'
text_data = data['Text'].values.tolist()

import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer

# Download required resources
nltk.download('stopwords')
nltk.download('punkt')

# Preprocess the text data
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()

preprocessed_data = []
for text in text_data:
    tokens = word_tokenize(text.lower())  # Tokenize the text
    filtered_tokens = [stemmer.stem(token) for token in tokens if token.isalpha() and token not in stop_words]  # Remove stop words and non-alphabetic tokens
    preprocessed_data.append(filtered_tokens)
from sklearn.feature_extraction.text import TfidfVectorizer

# Convert preprocessed data back to text
text_data_processed = [' '.join(tokens) for tokens in preprocessed_data]

# Create TF-IDF vectorizer
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(text_data_processed)

# Get the top keywords for each document
top_keywords = []
for i, doc in enumerate(text_data_processed):
    feature_values = tfidf_matrix[i, :].toarray()[0]
    top_keyword_indices = feature_values.argsort()[-5:][::-1]  # Get top 5 keywords (adjust as needed)
    top_keywords.append([vectorizer.get_feature_names()[index] for index in top_keyword_indices])

# Print the top keywords for each document
for i, doc_keywords in enumerate(top_keywords):
    print(f"Document {i + 1}: {doc_keywords}")
from gensim import corpora, models

# Create a dictionary from the preprocessed data
dictionary = corpora.Dictionary(preprocessed_data)

# Create a document-term matrix
corpus = [dictionary.doc2bow(tokens) for tokens in preprocessed_data]

# Train the LDA model
lda_model = models.LdaModel(corpus, num_topics=5, id2word=dictionary, passes=10)  # Adjust the number of topics as needed

# Print the topics
for topic in lda_model.print_topics():
    print(topic)
```
